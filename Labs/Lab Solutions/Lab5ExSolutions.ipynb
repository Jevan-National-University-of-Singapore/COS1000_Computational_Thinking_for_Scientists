# <span style = "color:blue">Question 1:</span>

Given the csv file SocialMedia.csv, plot a horizontal bar chart of the number of users of the various social media platforms in the year 2017. Label your chart clearly.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Prepare data
data = pd.read_csv('SocialMedia.csv')              # Return a dataframe from the csv file
data_2017 = data[data['Year']==2017]               # Obtain data for the year 2017
print(data_2017,"\n")
data_2017np = pd.DataFrame(data_2017).to_numpy()   # Convert dataframe to a numpy array

user, platform = data_2017np[:,3]/1000000, data_2017np[:,0]   # Assign numpy data in column 4 (convert to millions)
#print(user)                                                   # and column 0 to variables
#print(platform)

# Create plot
plt.figure(1,(10,7))
plt.barh(platform, user, color="orange")   

plt.title("Number of users of social media platforms in the year 2017", fontsize=15)   
plt.xlabel("Number of users (in millions)",fontsize=12)
plt.ylabel("Social media platform", fontsize=12)
plt.show()
```

# <span style = "color:blue">Question 2:</span>
Given the csv file on life expectancy, write a program that will chart the life expectancy of three countries of your choice. Your plot should show the period from 1950 to 2019 (inclusive).

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('LifeExpectancy.csv')   # Creating dataframe from csv file

df_ALG = data[data['Code'] == 'DZA']
data_ALG = df_ALG[(df_ALG['Year']>=1950) & (df_ALG['Year']<=2019)]   # Return a dataframe that contains data where logical operation returns true  
data_ALG = pd.DataFrame(data_ALG).to_numpy()

df_ARG = data[data['Code'] == 'ARG']
data_ARG = df_ARG[(df_ARG['Year']>=1950) & (df_ARG['Year']<=2019)]   # Return a dataframe that contains data where logical operation returns true  
data_ARG = pd.DataFrame(data_ARG).to_numpy()

df_AUS = data[data['Code'] == 'AUS']
data_AUS = df_AUS[(df_AUS['Year']>=1950) & (df_AUS['Year']<=2019)]   # Return a dataframe that contains data where logical operation returns true  
data_AUS = pd.DataFrame(data_AUS).to_numpy()

plt.figure(1,(10,6))
plt.plot(data_ALG[:,2],data_ALG[:,3],"g-",label="Algeria")     # The 3rd field is the year; 4th field is the life expectancy
plt.plot(data_ARG[:,2],data_ARG[:,3],"b-",label="Argentina")   
plt.plot(data_AUS[:,2],data_AUS[:,3],"r-",label="Australia")   

plt.title("Life expectancy from 1950 to 2019")
plt.ylabel("Life Expectancy"); plt.xlabel("Year")
plt.legend()
plt.show()
```
