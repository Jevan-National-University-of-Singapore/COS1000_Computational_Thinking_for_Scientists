## <span style = "color:blue">Question 1 Solutions</span>

```python
# Program to guess a number generated randomly

import random as ran     # Import random module to generate random numbers

num = ran.randint(0,9)   # randint() is a function to generate an integer, in this case between 0 and 9 inclusive
#print(num)               # Uncomment to see what number the program has generated

num_guess = 1

try:
    guess = eval(input("Guess an integer between 0 and 9 inclusive: "))
    
    while (type(guess) == int) and (0 <= guess < 10):
        if guess == num:
            print("Congratulations! You have guessed the number correctly!")
            print("Number of guesses you have used:",num_guess)
            break
        elif guess < num:
            guess = eval(input("Your number is less than the computer number. Please try again: "))
            num_guess += 1
        else:
            guess = eval(input("Your number is more than the computer number. Please try again: "))
            num_guess += 1
    else:
        print("Your entry is invalid! Goodbye!")
except:
    print("Your entry is invalid! Goodbye!")
```

## <span style = "color:blue">Question 2</span>
Write a program to print a multiplication table. Your program should prompt user to enter the number for the multiplication table. Using another cell, modify your previous program so that it could display all the multiplication tables up to the number.

## <span style = "color:blue">Question 2 Solutions</span>

```python
# Program to print a multiplication table

n = eval(input("Enter number for the multiplication table: "))

for x in range(10):  
    print(f"{x+1} x {n} = {(x+1)*n}")
```

```python
# Program to print all the multiplication tables up to the number

nMax = eval(input("Enter maximum number to display multiplication tables: "))

for j in range(1, nMax+1):   # Outer loop iterates nMax number of times
    for i in range(1, 11):   # Inner loop iterates 10 times
        print(f"{i} x {j} = {i*j}")
    print()      
```
