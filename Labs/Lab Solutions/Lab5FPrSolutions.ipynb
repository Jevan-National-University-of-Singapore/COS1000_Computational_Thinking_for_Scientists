### Question 1:
Plot the graphs of the following functions using a 2 by 2 layout. The first few lines of code have been written for you.

(i) $ f(x) = x^3 $

(ii) $ f(x) = x^2-1$

(iii) $ f(x) = \sin x - \cos^2 x $

(iv) $ f(x) = x^4-3x^3+2x^2-5 $

```python
# Load modules
import matplotlib.pyplot as plt  
import numpy as np               

x = np.linspace(-5.0,5.0,100)   # Create numpy x values for plotting

plt.figure(figsize=(12, 10))    # Create a figure object which is 12 inches in width and 10 inches in height
                             
# Plot 1st graph 
plt.subplot(2,2,1)   # Position at top left of 2 x 2 layout
plt.plot(x,x**3,"r-")
plt.title("$x^3$ vs $x$")
plt.xlabel("$x$"); plt.ylabel("$x^3$")   # Can be written as one line of code separated by ";"
plt.grid(True)

# Plot 2nd graph
plt.subplot(2,2,2)   # Position at top right of 2 x 2 layout
plt.plot(x,x**2-1,"b-")
plt.title("$x^2-1$ vs $x$")
plt.xlabel("$x$"); plt.ylabel("$x^2-1$")
plt.grid(True)

# Plot 3rd graph 
plt.subplot(2,2,3)   # Position at bottom left of 2 x 2 layout
plt.plot(x,np.sin(x)-np.cos(x)**2,"g-")
plt.title("$\sin x - \cos^2x$ vs $x$")
plt.xlabel("$x$"); plt.ylabel("$\sin x - \cos^2x$")
plt.grid(True)

# Plot 4th graph
plt.subplot(2,2,4)   # Position at bottom right of 2 x 2 layout
plt.plot(x,x**4-3*x**3+2*x**2-5,"m-")
plt.title("$x^4-3x^3+2x^2-5$ vs $x$")
plt.xlabel("$x$"); plt.ylabel("$x^4-3x^3+2x^2-5$")
plt.grid(True)

# Finishing the plots
plt.subplots_adjust(wspace=0.3)   # To avoid overlapping of the graphs
plt.subplots_adjust(hspace=0.3)
#plt.tight_layout()   # tight_layout() function also adjusts spacing (in a 'tight' manner) and could replace above two lines
plt.show()
```

### Question 2:
Given some data that show the carbon dioxide emissions (in million tons) of three countries from 2015 to 2019, write a program to plot the emissions over the same corresponding period of time.

```python
# Remember you not only need to be clever in plotting graphs but also in interpreting them!
# In this question, comparison of CO2 emission levels should be interpreted or read in the context of country and population 
# size and extent of industrialisation, etc.

import matplotlib.pyplot as plt
  
# Create data
x = ['2015', '2016', '2017', '2018', '2019']   # Years entered as string as integers will output decimal display in this case
y1 = [507, 568, 531, 577, 618]   # Emission corresponding to Indonesia
y2 = [233, 247, 249, 249, 250]   # Emission corresponding to Malaysia
y3 = [62, 40, 39, 38, 39]        # Emission corresponding to Singapore

plt.figure(figsize=(8,5))   # Set figure size
plt.plot(x, y1, "rx-", label="Indonesia")   # Set colour as well as line and marker styles for legend
plt.plot(x, y2, "bo--", label="Malaysia")
plt.plot(x, y3, "g^:", label="Singapore")

plt.title("Carbon dioxide emissions of Indonesia, Malaysia and Singapore")
plt.xlabel("Year")
plt.ylabel("CO$_2$ Emission (in million tons)")
plt.legend(loc=0)
#plt.legend(bbox_to_anchor=(1.0,0.8))   # Another way to code placement of legend: x and y co-ordinates of 
plt.show()                              # the upper right-hand corner of legend are specified  
                                        # (1.0, 1.0) will be the top right corner of plotting area
```

### Question 3:
Referring to books.csv in Example 8A, generate a csv file that contains data records of books with the genre 'Philosophy' and height that is less than 200 units.
