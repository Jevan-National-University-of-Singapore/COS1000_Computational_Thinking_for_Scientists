### Question 1:
Write a program that implements a function which will return the maximum value in the list of numeric values passed to it. Test your program with different lists of values.

```python
def findMax(x):
    m = x[0]
    for num in x:
        if num > m:
            m = num
    return m

list1 = [4,5,-2,7,0]
list2 = [23,-10,45,55,17]

"""
To understand the for loop in the function, look at how m and num change after each iteration when list1 is passed:

m = 4 (before loop)

iteration    num     m
First         4      4
Second        5      5
Third        -2      5
Fourth        7      7
Fifth         0      7

m = 7 (after loop exits which will be returned to function call code)

"""

print(f"Maximum value found in List 1 is {findMax(list1)}")
print(f"Maximum value found in List 2 is {findMax(list2)}")
```

### Question 2:
Write a program that implements a (recursive) function to sum all the positive odd integers starting from 1 to the maximum positive odd integer specified by user. Assuming only a valid integer is entered, there is no necessity to check for invalid entry in your program. The following shows one instance of running the program:


```python
# You can skip this since recursion code is optional in this module; 
# be careful not to enter an even number as code will result in infinite recursion (think about why
# this is the case) and you will need to interrupt and restart kernel.

def sumOdd(x):
    if x == 1:
        return 1
    else:
        return (x + sumOdd(x-2))

num = eval(input("Enter the maximum odd integer to sum to starting from 1: "))
print(f"The sum of all odd integers starting from 1 to {num} is {sumOdd(num)}")
```

```python
# Non-recursive function

def sumOdd(x):
    total = 0
    for i in range(1,x+1,2):
        total += i
    return total

num = eval(input("Enter the maximum odd integer to sum to starting from 1: "))
print(f"The sum of all odd integers starting from 1 to {num} is {sumOdd(num)}")
```

### Question 3:
Write a program that uses lambda statements to implement the following mathematical functions: 

&emsp;&nbsp;(i) $f(x,y) = x\cos y - 3x $

&emsp;&nbsp;(ii) $f(x,y,z) = 2x^3 + 5xz^2 + 3\sqrt{y}z$

You are to prompt user to enter a value for x, y and z. Assuming only valid numeric values are entered (integers or floats), there is no necessity to check for invalid entry in your program. Round your results to 2 decimal places. The following shows one instance of running the program:


```python
import math as m

x = eval(input("Enter a value for x: "))
y = eval(input("Enter a value for y: "))
z = eval(input("Enter a value for z: "))

func1 = lambda x, y: x*m.cos(y)-3*x
print("Function 1 evaluates to:",round(func1(x,y),2))
#print(f"Function 1 evaluates to: {func1(x,y):.2f}")   # Another way to output using f string

func2 = lambda x, y, z: 2*x**3 + 5*x*z**2 + 3*m.sqrt(y)*z
print("Function 2 evaluates to:",round(func2(x,y,z),2))
#print(f"Function 2 evaluates to: {func2(x,y,z):.2f}")   # Another way to output using f string
```
