## <span style="color:blue">Line plot</span>
A [line](https://www.storytellingwithdata.com/blog/2020/3/24/what-is-a-line-graph) plot displays data along a number line or over a time series. It is an excellent choice when the purpose is to show how the value of a variable varies or to track how several such values compare to one another over time.

```python
# Example 1A: A single Line plot

import matplotlib.pyplot as plt

x = [1600, 1700, 1800, 1900, 2000]   # Year data or x-coordinates
y = [0.2, 0.5, 1.1, 2.2, 7.7]        # Population in billions or y-coordinates

plt.plot(x,y)   # x and y are parameters corresponding to the horizontal and vertical coordinates of the data points
                # x values are optional; if absent, a default will be generated [0., 1., 2., 3., 4.] 
plt.title("World Population over Time")   # Set title for plot
plt.xlabel("Year")                        # Set label for x axis
plt.ylabel("Population (in billions)")    # Set label for y axis
plt.show()   # Render the plot on screen

```

## <span style="color:blue">Line and marker styles and colours</span>
The line and marker styles are specified by character codes in single or double quotes (for example, <span style="color:blue">"ro-"</span>). These characters are actually shorthand for much longer expressions. For example,

    plt.plot(x,np.sin(x),"^-b")

is equivalent to the following:

    plt.plot(x,np.sin(x), marker="^", linestyle="-", color="b")


For efficiency, these settings are usually condensed and in the order of marker, line and colour (or color, marker and line). Some common line and marker styles and their corresponding specifiers are shown in the table below:

|'o' | Circle marker |

You can also specify the colour of the lines/markers. The following is a list of codes for specifying the colours:

## <span style="color:blue">Displaying multiple plots</span>
It is possible to have more than one plot in a figure layout. You can use the command 

    plt.subplot(rows, cols, plot number)

to establish a subplot window within the current figure. The parameters *rows* and *cols* divide the figure into *rows* $\times$ *cols* grid of subplots.

```python
# Example 2: Creating two plots using a subplot layout

import matplotlib.pyplot as plt  
import numpy as np

x = np.arange(-6.3,6.4,0.2)   # Create numpy array of floats for plotting
#x = np.arange(-2*np.pi, 2*np.pi, np.pi/10)

plt.figure(figsize=(6,8))   # Create a figure object which is 6 inches in width and 8 inches in height.
                            # Each unit inch is 80 x 80 pixels; default is (6.4,4.8) 
# Plot 1st graph 
plt.subplot(2,1,1)   # Position at top of 2 x 1 layout
plt.plot(x,np.sin(x),"r-")
plt.title("Graph of sine function")
plt.xlabel("x"); plt.ylabel("sin(x)")   # Can be written as one line of code separated by ";"
plt.grid(True)

# Plot 2nd graph
plt.subplot(2,1,2)   # Position at bottom of 2 x 1 layout
plt.plot(x,np.cos(x),"b-")
plt.title("Graph of cosine function")
plt.xlabel("x"); plt.ylabel("cos(x)")
plt.grid(True)

# Finishing the plots
plt.subplots_adjust(hspace=0.3)   # To avoid overlapping of the graphs vertically
plt.show()
```

## <span style="color:blue">Error bars</span>
Error bars in a plot indicate the estimated error or uncertainty to give a general sense of how precise a measurement is.

```python
# Example 3 Plotting with error bars

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(11)/10   # arange(11) --> [0 1...10]; arange(11)/10 --> [0/10  1/10  2/10...10/10]; 
#x = np.arange(0,1.1,0.1)   # Another way to code 
#x = np.linspace(0,1,11)    # Another way to code
print(x) 

y = (x + 0.1)**3       # y is dependent on x and is also a numpy array object
print(y)
 
# Defining the y error; y_error = 0.05
y_error = 0.05
                                    
plt.plot(x,y)
plt.errorbar(x, y, yerr = y_error, fmt ='o')
plt.title("Plot of y vs x with error bars")
plt.xlabel("x"); plt.ylabel("y")
plt.show()
```

## <span style="color:blue">Log Plot</span>
A [semi log plot](https://pythontic.com/visualization/charts/semilog) is a graph where the data on one axis is on a logarithmic scale (either y axis or x axis) and the data on the other axis is on a normal (or linear) scale. On a semi-log plot the [spacing](https://en.wikipedia.org/wiki/Semi-log_plot) of the scale on the y-axis (or x-axis) is proportional to the logarithm of the number, not the number itself. It is equivalent to converting the y values (or x values) to their log, and plotting the data on linear scales. An example of an exponential growth would be [bacterial growth](https://en.wikipedia.org/wiki/Bacterial_growth).

```python
# Example 4 Plotting log graphs

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(12,10))

plt.subplot(2,2,1)

x1 = np.arange(51)   # Create a numpy array of 51 x data 
print(type(x1))            
print("x1 =",x1)

y1 = [2**x for x in range(51)]   # Create a list of y data that grows exponentially as 2^x
#print(type(y1))
#print("y1 =",y1)

plt.plot(x1, y1)    
plt.title("y vs x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)   
              
plt.subplot(2,2,2)
plt.semilogy(x1,y1)   # Plot y data on a base-10 logarithmic scale
#plt.yscale("log")    # Convert y-axis to base-10 Logarithmic scale; another way to code
#plt.plot(y1)
plt.title("y vs x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)   

plt.subplot(2,2,3)
y2 = [np.e**x for x in range(51)]   # Create a list of y data that grows exponentially as e^x
#print(type(y2))
#print("y2 =",y2)                       
plt.plot(x1,y2)
plt.title("y vs x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)   
 
plt.subplot(2,2,4)
y3 = np.log(y2)   # Create a numpy array of y data where y = log(base e)(y2) (y2 = e^x)
#print(type(y3))
#print("y3 =",y3)
plt.plot(x1,y3)
plt.title("y vs x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)   

plt.subplots_adjust(hspace=0.3)   # To avoid overlapping of the graphs vertically
plt.show()
```

## <span style="color:blue">Bar plot</span>
[Bar](https://www.storytellingwithdata.com/blog/2020/2/19/what-is-a-bar-chart) plot is a common chart type for plotting data by categories or data sorted into groups. It conveys relational information but changes over time can also be tracked. Data is encoded by the length of the bar. An important rule to follow when using bar plot is to use a zero baseline.

```python
# Example 5A: Vertical bar plot

import matplotlib.pyplot as plt   

num_hseholds, pets = [20, 36, 24, 12, 8], ["Cat", "Dog", "Hamster", "Rabbit", "Terrapin"]   # Initialise two lists

plt.figure(figsize=(7, 5))
#plt.bar(pets, num_hseholds)
plt.bar(pets, num_hseholds, color="orange", width = 0.5)   # Default width = 0.8

plt.title("Favourite Pets in Singapore", fontsize=15)   
plt.xlabel("Pets",fontsize=12)
plt.ylabel("Number of households in neighbourhood", fontsize=12)
plt.show()
```

```python
# Example 5B: Horizontal bar plot

import matplotlib.pyplot as plt   

num_hseholds, pets = [20, 36, 24, 12, 8], ["Cat", "Dog", "Hamster", "Rabbit", "Terrapin"]   # Initialise two lists

plt.figure(figsize=(7, 5))
plt.barh(pets, num_hseholds, color="orange", height=0.5)   # height in barh is equivalent to width in bar

plt.title("Favourite Pets in Singapore", fontsize=15)   
plt.xlabel("Pets",fontsize=12)
plt.ylabel("Number of households in neighbourhood", fontsize=12)
plt.xticks(np.arange(5,45,5))   # xticks create tick marks on x axis
plt.show()
```

```python
# Example 5C: Grouped (clustered) bar plot

import matplotlib.pyplot as plt
import numpy as np

# Create data
x = np.arange(5)   
y1 = [45, 56, 32, 89, 67]
y2 = [32, 56, 78, 48, 90]
y3 = [24, 37, 45, 55, 89]
width = 0.25
  
plt.figure(figsize=(8,5))
    
# Plot data by groups of three vertical bars
plt.bar(x-0.25, y1, width, color="pink")   # Placement of bar is offset by width
plt.bar(x, y2, width, color='green')
plt.bar(x+0.25, y3, width, color='orange')

plt.title("Team Scores in Three Rounds", fontsize=18)   
plt.xticks(x, ['Team A', 'Team B', 'Team C', 'Team D', 'Team E'])   # Convert value(s) to axis labels
plt.xlabel("Teams",fontsize=14)
plt.ylabel("Scores",fontsize=14)
plt.legend(["Round 1", "Round 2", "Round 3"])
plt.show()
```

## <span style="color:blue">Stack plot</span>
[Stack](https://chartio.com/learn/charts/stacked-bar-chart-complete-guide/) plot creates a stacked area plot which displays various components of data or parts to the whole. It shows how a larger category is divided into smaller categories and the relationship each part has to the total. Stacked bar plot does the same but the challenge with stacked bars is that it is hard to compare the subcomponents across categories that are not aligned to the same baseline. 

```python
# Example 6A: Stack plot

import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
  
exp_food = [300, 250, 320, 300, 350, 550]
exp_transport = [200, 180, 220, 230, 240, 450]
exp_utilities = [150, 165, 140, 145, 150, 125]
  
# Stackplot with different colour values
plt.stackplot(months, exp_food, exp_transport, exp_utilities, colors=['pink','green','orange'])

plt.title("Expenses by month (2022)")
plt.xlabel("Month")
plt.ylabel("Amount (in dollars)")
plt.legend(("Food","Transport","Utilities"),loc=2) 
plt.show()
```

```python
# Example 6B: Stack plot (Optional); demonstrating the use of dictionaries

import matplotlib.pyplot as plt
import numpy as np

year = [1835, 1950, 1960, 1980, 1990, 2000, 2007, 2021]   # Declare years in a list

# Declare a dictionary of sales of brands ordered by year declared in list
sales = {
    'Brand 1': [428, 1284, 365, 477, 631, 814, 1044, 1275],        
    'Brand 2': [340, 425, 519, 619, 727, 840, 943, 1006],
    'Brand 3': [1394, 1686, 2120, 2625, 3202, 3714, 4169, 4560],
    'Brand 4': [220, 253, 276, 295, 310, 303, 294, 293]
}                                                                    
       
plt.stackplot(year, sales.values(), labels=sales.keys())

plt.legend(loc='upper left')   # Specify loc by text
plt.title('Sales by brand through the years')
plt.xlabel('Year')
plt.ylabel('Number of sales')
plt.show()
```

```python
# Example 6C: Stack bar plot; notice it is hard to compare subcomponents across categroies 

import matplotlib.pyplot as plt
import numpy as np
  
# Create data
x = ['Sep', 'Oct', 'Nov', 'Dec']
y1 = np.array([36, 28, 30, 48])   # car
y2 = np.array([22, 23, 22, 30])   # motorcyle
y3 = np.array([6, 5, 7, 8])       # lorry

plt.figure(figsize=(6,5))

# Plot bars in stack manner
plt.bar(x, y1, color='orange', width=0.5)
plt.bar(x, y2, bottom=y1, color='green', width=0.5)
plt.bar(x, y3, bottom=y1+y2, color='maroon',width=0.5)
plt.xlabel("Months")
plt.ylabel("Number of vehicles")
plt.legend(("Car", "Motorcylce", "Lorry"))
plt.title("Number of accidents by vehicle type in 2021")
plt.show()
```

## <span style="color:blue">Scatter plot</span>
[Scatter](https://www.storytellingwithdata.com/blog/2020/5/27/what-is-a-scatterplot) plots are used to elucidate relationship between variables (if any) and use dots to represent them. They are common in scientific fields and often used to understand data rather than to communicate with it.

```python
# Example 7A: A scatter plot; dataset 2 demonstrates a strong correlation between the variables but not dataset 1

import matplotlib.pyplot as plt

# Declare dataset 1
x1 = [89, 43, 36, 36, 95, 10, 66, 34, 38, 20]
y1 = [21, 46, 3, 35, 67, 95, 53, 72, 58, 10]
  
# Declare dataset 2
x2 = [16, 25, 32, 43, 50, 62, 78, 83, 90, 97]
y2 = [12, 29, 35, 47, 54, 60, 75, 82, 89, 95]
  
plt.scatter(x1, y1, c="pink", linewidths=2, marker="s", edgecolor="green", s=50)  # s is marker size
plt.scatter(x2, y2, c="yellow", linewidths=2, marker="^", edgecolor="red", s=200)

plt.title("y vs x")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(("Dataset 1","Dataset 2"),loc=2) 
plt.show()
```

```python
# Example 7B: A scatter plot using random numbers

import matplotlib.pyplot as plt
import numpy as np

np.random.seed(2)   # Comment this if you do not wish to fix the randomness to a particular seed

# Generates random data; n = number of points & x, y are their coordinates
n = 50
x = np.random.rand(n)   # Generate n numbers within [0,1) randomly; x is a numpy array 
print(x)
y = x + 3*np.sin(np.random.rand(n) - 0.5)   # y is also a numpy array
print(y)
        
# Randomly determine the size of data points
size = 200.0 * np.random.rand(n)

# c controls color, alpha controls transparency
# For dots that appear darker, it means there has been overlapping
plt.scatter(x, y, c="green", s=size, alpha=0.32, edgecolors='none')  

plt.title("y vs x")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)   # To render grid lines
plt.show()
```
