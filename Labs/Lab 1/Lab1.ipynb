## <span style="color:blue">Entering mathematical expressions with LaTeX</span>

In communicating scientific ideas, mathematical expressions are essential. Most of you have probably written equations in mainstream authoring tools like Microsoft Word. While intuitive, they quickly become frustrating when more complicated expressions are required. The following shows how arbitrarily complex mathematical expressions can be composed in Jupyter Notebook via LaTeX, another document preparation system.
- LaTeX is an extensive typesetting system for scientific documentation that supports specialised symbols for mathematics, statistics, physics, computer science and more. 
- It is beyond the scope of this lab course to cover everything, only the basics of writing high quality mathematical equations will be covered.
- When a Markdown cell is executed, the LaTeX portions are automatically rendered as equations with high quality typography.
- Within a Markdown cell, we can include mathematical expressions with sandwiched LaTeX codes
- For inline equations, single dollar signs are used: $1 + 2 = 3$
- For equations displayed on their own, double dollar signs are used: $$1+2=3$$

### <span style="color:blue">Fractions</span>

Fractions can be expressed in the following ways:

- <span style="color:blue">^1/_2</span> <br></br><br>
- <span style="color:blue">\frac{1}{2x}</span> <br></br><br></br>
- <span style="color:blue">3 \over 4</span>

Their rendered LaTeX output are $^1/_2$, $\frac1{2x}$ and $3 \over 4$ respectively.

### <span style="color:blue">Roots</span>

Examine the example below:

<center><span style="color:blue">f(x) = \sqrt[3]{2x} + \sqrt{x-2}</span></center>

When rendered, it reads $f(x) = \sqrt[3]{2x} + \sqrt{x-2}$. In general, the n-th root of m can be coded as \sqrt[n]{m}. Take note of the use of square and curly brackets.

### <span style="color:blue">Superscripts and subscripts</span>

We can use the ^ (hat) and _ (underscore) symbols for implementing superscripts and subscripts. Use curly brackets if the super/subscript contains more than one character.

<center><span style="color:blue">x^2 </span></center>
<center><span style="color:blue">x_1 + x_2+x_3 </span></center>
<center><span style="color:blue">x^{x^{x^{-x}}} </span></center>



These render as $x^2$, $x_1+x_2+x_3$ and $x^{x^{x^{-x}}} $ respectively. Note that the brackets matter, even when the order of super/subscripts do not:

<center><span style="color:blue">{y_3}^2 + y_3^2+y^2_3 + y_{3^2} </span></center>

renders as ${y_3}^2 + y_3^2+y^2_3 + y_{3^2}$.

### <span style="color:blue">Log and trigo functions</span>

<center><span style="color:blue">\sin x + \cos x + \tan x </span></center>
<center><span style="color:blue">\log x + \cosh y </span></center>
<center><span style="color:blue">\sin(\log_{10}x)  </span></center>

These render as $\sin x + \cos x + \tan x $, $\log x+\cosh y$ and $\sin(\log_{10}x)$ respectively.

### <span style="color:blue">Sums, products and calculus</span>

<center><span style="color:blue">\sum\limits_{a}^{b}X^a</span></center>
<center><span style="color:blue">\prod\limits_{a=1}^{b}Y_a</span></center> 
<center><span style="color:blue">\int_a^b \frac{\partial f}{\partial x}dx </span></center>
<center><span style="color:blue"> \lim\limits_{x \to 0} \frac{\sin x}{x}=1</span></center>
    
which render as $\sum\limits_{a}^{b}X^a$, $\prod\limits_{a=1}^{b}Y_a$, $\int_a^b \frac{\partial f}{\partial x}dx$ and $\lim\limits_{x \to 0} \frac{\sin x}{x}=1$ respectively. The \limits command deserves some clarification. It is **not** the limit operator, but is rather an instruction to keep the subscript directly underneath the symbol. The limit operator is given by \lim, as demonstrated in the fourth example.

### <span style="color:blue">Brackets</span>

There are four types of brackets:
- parentheses or "round brackets" ( )
- "box brackets" or "square brackets" [ ]
- "braces" or "curly brackets" { }
- "angle brackets" < >

They can be used freely in mathematical expressions. In more complicated expressions, it may be aesthetically desirable to make certain brackets larger. This can be done through the \left and \right commands: see the difference between <span style="color:blue"> (\sqrt{\frac{x}{y}}) </span> vs <span style="color:blue"> \left(\sqrt{\frac{x}{y}}\right) </span> i.e. $(\sqrt{\frac{x}{y}})$ vs $\left(\sqrt{\frac{x}{y}}\right)$? Please remember to close the bracket with a \right if you have opened it with a \left !
