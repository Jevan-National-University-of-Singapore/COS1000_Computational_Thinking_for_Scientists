Write a program to evaluate the following arithmetic expressions. For (b) to (d), prompt for user input of the value for variable $x$. Consider using a new cell for each part and output the result using a print statement. There is no necessity to format the number of decimal places.

(a) result = $(15+23.4) \times 0.25 \div 18.92 - 3.167$

(b) result = $ 2x^4 + 3x^3 - 4x^2 + 5x - 10$

(c) result = $ \dfrac{1}{x} + \dfrac{2}{x^2} + \dfrac{3}{x^3} $

(d) result = $ \dfrac{\left(\dfrac{x^3}{5} - 3x\right)}{7-2x} + 2^x - 1 $

```python
def question_a(x:float)->float:
    return (15+23.4)*0.25/18.92-3.167

def question_b(x:float)->float:
    return 2*(x**4) + 3*(x**3) - 4*(x**2) + (5*x) - 10

def question_c(x:float)->float:
    assert x != 0
    return (1/x)+(2/(x**2)) + (3/(x**3))
    
def question_d(x:float)->float:
    assert x != 3.5
    return ((((x**3)/5)-(3*x))/(7-(2*x))) + 2**x - 1

questions = [question_a, question_b, question_c, question_d]

for question in questions:
    print(
        question(
            float(input("value of x:"))
        )
    )
```
